```python
import numpy as np
import matplotlib.pyplot as plt

# Days of the year
N = np.arange(1, 366)

# Calculate declination
delta = -23.44 * np.cos(np.deg2rad((360 / 365) * (N + 10)))

# Calculate rate of change
d_delta = (360 / 365) * 23.44 * np.sin(np.deg2rad((360 / 365) * (N + 10)))

# Plot declination
plt.figure(figsize=(12, 6))
plt.plot(N, delta, label="Sun's Declination")
plt.plot(N, d_delta, label="Rate of Change", linestyle='--')
plt.title("Sun's Declination and Rate of Change Over the Year")
plt.xlabel("Day of the Year")
plt.ylabel("Degrees")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("suns_declination.png")  # Saves the figure as a PNG file
plt.show()
```
