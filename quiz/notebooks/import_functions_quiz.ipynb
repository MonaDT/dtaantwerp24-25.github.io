### What does this code do?

```python
with open('poirot.txt', 'r') as f:
    text_lines = f.readlines()
    
def is_vowel (letter):
    'Test whether letter is a vowel'
    return letter in 'aeiou'

def cv_structure (word):
    'Give the cv structure of a word'
    result = ""
    for letter in word:
        if is_vowel (letter):
            result += 'V'
        else:
            result += 'C'
    return result

for line in text_lines:
    for w in line.split():
        w = w.lower()
        if is_vowel(w[0]):
            print(cv_structure(w))
        else:
            print(cv_structure(w))
```
